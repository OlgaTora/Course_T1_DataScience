Создайте функцию , которая будет применена к столбцу рейтинг методом `apply()`.
Функция должна принимать строку с рейтингом и возвращать:
 - High rating, если рейтинг больше или равен 4.5
 - Middle rating, если рейтинг между 3.8 и 4.5
 - Low rating, если рейтинг меньше 3.8 
Результат работы функции запишите в новый столбец с названием `categorical_rating`

```python
def get_rating(rating: int) -> str:
    if rating >= 4.5:
        return 'High rating'
    elif rating < 3.8:
        return 'Low rating'
    else:
       return 'Middle rating'
```

```python
df['categorical_rating'] = df.rating.apply(get_rating)
df.head(2)
```

Изучите функцию `transform` в документации Pandas. Примените её к датафрейму, который будет заранее сгруппирован по столбцу `category`. Создайте функцию , которая будет применена к столбцу `last updated` и возвращать из него число месяца, если число месяца вернуть из строки невозможно, то верните вместо этого строку с значением `miss_date`.\
Запишите полученный столбец в датафрейм , название столбцу дайте `day_of_update`.

Если вы столкнётесь с некоторой ошибкой , которую вам выдаст Pandas, то изучите конструкцию Python `try` \ `except`.\
Вот псевдокод , который поможет лучше разобраться с этой конструкцией:

```python
from datetime import datetime
def get_month(some_date: str) -> int | str:
    try: 
        some_date = pd.to_datetime(some_date)
        return some_date.month
    except ValueError:
        return 'miss date'

df['day_of_update'] = df['last updated'].apply(get_month)
df.sample(3)
```

```python
df.day_of_update.value_counts()
```
